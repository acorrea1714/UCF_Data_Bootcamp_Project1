Import dependencies

```python
import pandas as pd
import json
import os
import matplotlib.pyplot as plt
```

Import Excel workbook using pandas

```python
filepath = os.path.join('.', 'Resources', 'SourceData', 'Revised_finc01_3.xlsx')
df = pd.read_excel (filepath, sheet_name = 'Type of Family', nrows=42, usecols="A,C:G")
df
```

```python
all_primary_families = df.iloc[1:, :2]
all_primary_families
```

```python
all_primary_families["Characteristic"] = all_primary_families["Characteristic"].str.replace('$','')
all_primary_families["Characteristic"] = all_primary_families["Characteristic"].str.replace(',','')
all_primary_families
```

```python
plt.rcParams.update({'figure.autolayout': True})

fig,ax = plt.subplots(figsize=(9,12))
ax.barh(all_primary_families["Characteristic"], all_primary_families["All primary families"])
ax.set(xlabel='Number of Households (Thousands)', ylabel='Median Income (Dollars)', title='Number of Households per Income Range (White Only)')
plt.show()

```
